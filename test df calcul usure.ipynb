```python
aero = pd.read_csv('aeronefs_2023-11-07.csv')
vol = pd.read_csv('logs_vols_2023-11-07.csv')
deg = pd.read_csv('degradations_2023-11-07.csv')
comp = pd.read_csv('composants_2023-11-07.csv')
```
